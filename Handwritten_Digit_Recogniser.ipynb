Handwritten digit Recognizer using MNIST Dataset. 

```python
import numpy as np
import pandas as pd
import tensorflow as tf 
from tensorflow import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import linear, relu, sigmoid
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
#saving the X_train and Y_train data
ans = "label"
X_train = df[df.columns.drop(ans)]
Y_train = df[[ans]].copy()
print("The X Training set:")
print(X_train)
print("\n")
print("The Y Training set:")
print(Y_train)
```

```python
print("Shape of X training set: ", X_train.shape)
print("Shape of Y training set: ", Y_train.shape)
```

```python
tf.random.set_seed(1234)
model = Sequential([
    tf.keras.Input(shape=(784,)),
    Dense(units=25, activation='relu', name="layer1"),
    Dense(units=15, activation='relu', name="layer2"),
    Dense(units=10, activation='linear', name="layer3"),
  ], name = "my_model"
)
```

```python
model.summary()
```

```python
[l1, l2, l3] = model.layers
```

```python
W1, b1 = l1.get_weights()
W2, b2 = l2.get_weights()
W3, b3 = l3.get_weights()
print(f"Variable\t shape\t\t Variable\t shape\t")
print(f"W1\t\t {W1.shape}\t b1\t\t {b1.shape} \t")
print(f"W2\t\t {W2.shape}\t b2\t\t {b2.shape} \t")
print(f"W3\t\t {W3.shape}\t b3\t\t {b3.shape} \t")
```

```python
model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True), 
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)

history = model.fit(
    X_train,Y_train,
    epochs=30
)
```

```python
X_test = pd.read_csv('test.csv')
Y_test = model.predict(X_test)
print("Shape of X_test: ", X_test.shape)
print("Shape of Y_test: ", Y_test.shape)
```

```python
print(X_test)
```

```python
result = pd.DataFrame({'ImageId':[], 'label': []})
for i in range(0, len(Y_test)):
  id = 1+i
  label = np.argmax(Y_test[i]) 
  tmp = pd.DataFrame({'ImageId':[id], 'label': [label]})
  result = result.append(tmp, ignore_index=True)
```

```python
result['ImageId'] = result['ImageId'].astype(int)
result['label'] = result['label'].astype(int)
result.to_csv('output.csv', index=False)
result.head(10)
```
